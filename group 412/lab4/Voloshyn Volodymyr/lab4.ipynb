```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
yolo  = cv2.dnn.readNetFromDarknet("yolov2.cfg", "yolov2.weights")
```

```python
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
```

```python
classes
```

```python
layer_names = yolo.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in yolo.getUnconnectedOutLayers()]
```

```python
cat = "cat.jpg"
traffic = "traffic.jpg"
```

```python
border_color = (0, 255, 0)
text_color = (255, 0, 100)

def detectinon(image_path):
  img = cv2.imread(image_path)
  height, width, channels = img.shape

  blob = cv2.dnn.blobFromImage(img, 1/255.0, (416, 416), swapRB=True, crop=False)

  yolo.setInput(blob)
  outputs = yolo.forward(output_layers)

  class_ids = []
  confidences = []
  boxes = []
  for output in outputs:
      for detection in output:
          scores = detection[5:]
          class_id = np.argmax(scores)
          confidence = scores[class_id]
          if confidence > 0.5:
              center_x = int(detection[0] * width)
              center_y = int(detection[1] * height)
              w = int(detection[2] * width)
              h = int(detection[3] * height)
              x = int(center_x - w / 2)
              y = int(center_y - h / 2)

              boxes.append([x, y, w, h])
              confidences.append(float(confidence))
              class_ids.append(class_id)

  indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

  for i in range(len(boxes)):
      if i in indexes:
          x, y, w, h = boxes[i]
          label = str(classes[class_ids[i]]) + " " + str(round(confidences[i], 3))
          cv2.rectangle(img, (x, y), (x + w, y + h), border_color, 3)
          cv2.putText(img, label, (x, y), cv2.FONT_HERSHEY_PLAIN, 3, text_color, 6)

  return img
```

```python
plt.imshow(detectinon(cat))
```

```python
plt.imshow(detectinon(traffic))
```
