```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# завантаження зображення і його конвертування в масив
image = Image.open("././images/im1.jpeg")
image_array = np.array(image)
```

```python
# виведення оригінального зображення та його гістограми
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image_array)
ax1.set_title("Оригінальне зображення")
ax2.hist(image_array.ravel(), bins=256)
ax2.set_title("Гістограма оригінального зображення")
plt.show()
```

```python
# виконання гістограмного зсуву ліворуч
shifted_image_array = image_array - 50
shifted_image_array[shifted_image_array < 0] = 0
```

```python
# виведення зсунутого зображення та його гістограми
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(shifted_image_array)
ax1.set_title("Зсунуте зображення")
ax2.hist(shifted_image_array.ravel(), bins=256)
ax2.set_title("Гістограма зсунутого зображення")
plt.show()
```
