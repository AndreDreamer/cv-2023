```python
import copy
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def filter2d(image_array, kernel_array):

    result = np.zeros_like(image_array)
    copy_image = copy.deepcopy(image_array)

    for i in range(kernel_array.shape[0]+1):
        copy_image = np.pad(copy_image, [(0, 1), (0, 1)], mode='constant')

    for x in range(image.shape[0]):
        for y in range(image.shape[1]):
            result[x][y] = (
                np.array(
                    copy_image[
                        x: (x+kernel_array.shape[0]),
                        y: (y+kernel_array.shape[1])
                    ]
                ) * np.array(kernel_array)
            ).sum()

    return result
```

```python

image = cv2.imread('../images/image-3.jpg', 0)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
kernel_x = np.array([
    [0, -1],
    [1, 0]]
)
kernel_y = np.array([
    [-1, 0],
    [0, 1]]
)

fig, axes = plt.subplots(1, 2, figsize=(10, 10))

im_r = filter2d(image_array=image, kernel_array=kernel_x)
axes[0].imshow(cv2.cvtColor(im_r, cv2.COLOR_BGR2RGB))
axes[0].set_title('Roberts - X')
axes[0].axis("off")

im_r = filter2d(image_array=image, kernel_array=kernel_y)
axes[1].imshow(cv2.cvtColor(im_r, cv2.COLOR_BGR2RGB))
axes[1].set_title('Roberts - Y')
axes[1].axis("off")

fig.show()
```
