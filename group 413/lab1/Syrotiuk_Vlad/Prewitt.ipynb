```python
from PIL import Image
import numpy as np
```

```python
image = Image.open("7.jpg")
```

```python
image
```

```python
gray_img = image.convert('L')
```

```python
img_array = np.array(gray_img)
```

```python
kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
```

```python
edge_img = np.zeros_like(img_array)
```

```python
for i in range(1, img_array.shape[0] - 1):
    for j in range(1, img_array.shape[1] - 1):
        grad_x = np.sum(kernel_x * img_array[i-1:i+2, j-1:j+2])
        grad_y = np.sum(kernel_y * img_array[i-1:i+2, j-1:j+2])
        
        edge_img[i, j] = np.sqrt(grad_x**2 + grad_y**2)

edge_img = Image.fromarray(edge_img.astype(np.uint8))
```
