```python
!pip install opencv-python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hist_shift (image, k):
    def shift(value):
        return max(0, min(value - k, 255))
    return np.vectorize(shift)(image).astype(np.uint8)
```

```python
def process_image(img_path):
    k_min, k_max, step = 0, 120, 40
    images = [hist_shift(cv2.imread(img_path), k) for k in range(k_min, k_max, step)]
    images = [cv2.resize(img, (300,500)) for img in images]

    k = [k for k in range(k_min, k_max, step)]
    for i in range(3):
        plt.hist(images[i].ravel(),256,[0,256])
        plt.title(f'k = {k[i]}')
        plt.show()

    images = np.concatenate(images, axis=1)
    cv2.imshow('image', images)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
images = ['cat1.jfif', 'cat2.jfif', 'cat3.jfif']
process_image(images[0])
```
