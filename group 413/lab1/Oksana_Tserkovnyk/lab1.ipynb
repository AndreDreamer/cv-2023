```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from PIL import Image
```

```python
low_det = np.array(Image.open('low_det-Copy1.jpg').convert('L'))
```

```python
def D1filter(image,k,l):
    img = image * k + l
    img[img > 255] = 255
    img[img < 0] = 0
    return img

def show_img(img):
    plt.figure(figsize=(20, 10))
    plt.imshow(img,  cmap="gray")
```

```python
show_img(low_det)
```

```python
show_img(D1filter(low_det, 2, 5))
```

```python
D1filter(low_det, 0.5, 5)
```
