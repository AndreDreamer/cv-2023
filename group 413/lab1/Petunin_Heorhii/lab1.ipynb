```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

![Formula](./formula.jpg)

```python
def leftShiftFilter(image: np.ndarray, K: int): #k max = 255
    result = image.copy() 

    for x in range(0, image.shape[0]):
        for y in range(0, image.shape[1]):
            for z in range(0, image.shape[2]):
                result[x][y][z] = min([image[x][y][z] + K, 255])
    return result
```

```python
def showFilter(img, k):
    plt.imshow(img)
    plt.axis('off')
    plt.title("Left shifk with k = " + str(k))
    plt.show()
```

```python
def showCommon(img):
    plt.imshow(img)
    plt.axis('off')
    plt.title("Default image")
    plt.show()
```

```python
imgContrast = cv2.cvtColor(cv2.imread('contrast.jpg'), cv2.COLOR_BGR2RGB)
imgNonContrast = cv2.cvtColor(cv2.imread('non_contrast.jpg'), cv2.COLOR_BGR2RGB)
imgDetalized = cv2.cvtColor(cv2.imread('detalized.jpg'), cv2.COLOR_BGR2RGB)
imgNonDetalized = cv2.cvtColor(cv2.imread('non_detalaized.jpg'), cv2.COLOR_BGR2RGB)
```

```python
showCommon(imgContrast)
shift = leftShiftFilter(imgContrast, 40)
showFilter(shift, 40)
```

```python
showCommon(imgNonContrast)
shift = leftShiftFilter(imgNonContrast, 40)
showFilter(shift, 40)
```

```python
showCommon(imgDetalized)
shift = leftShiftFilter(imgDetalized, 40)
showFilter(shift, 40)
```

```python
showCommon(imgNonDetalized)
shift = leftShiftFilter(imgNonDetalized, 40)
showFilter(shift, 40)
```
