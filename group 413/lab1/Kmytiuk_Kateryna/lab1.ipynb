```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
imgs = [plt.imread(f) for f in ['lab1_1.jpg', 'lab1_2.jpg', 'lab1_3.jpg', 'lab1_4.jpg']]
```

```python
def gamma_increase(img, gamma):
    return np.uint8(255 * np.power(img / 255.0, 1 / gamma))
```

```python
def show(img, img_gamma, gamma):
    _, axs = plt.subplots(2, 3, figsize=(20, 5))

    axs[0].imshow(img)
    axs[0].set_title('Original')
    axs[0].axis('off')
    
    axs[1].hist(img.ravel(), bins=256, range=(0, 255))
    axs[1].set_title('Original hist')
    
    axs[2].imshow(img_gamma)
    axs[2].set_title('Gamma corrected (gamma = ' + str(gamma) + ')')
    axs[2].axis('off')
    
    axs[3].hist(img_gamma.ravel(), bins=256, range=(0, 255))
    axs[3].set_title('Corrected hist')
    
    plt.show()
```

```python
gamma = 2.5
for img in imgs:
    show(img, gamma_increase(img, gamma), gamma)
```
