```python
x, img = data.transforms.presets.ssd.load_test(image, short=512)
plt.imshow(img)
class_IDs, scores, bounding_boxes = net(x)

ax = utils.viz.plot_bbox(img, bounding_boxes[0], scores[0],
                        class_IDs[0], class_names=net.classes)
plt.show()
```
