```python
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np


drive.mount("/content/drive")
```

```python
def gamma_increase(img, gamma):
    return (255 * np.power(img / 255, 1 / gamma)).astype(np.uint8)

def plot_images(imgs, imgs_gamma, labels):
    length = len(imgs)

    fig, axs = plt.subplots(length, 2, figsize=(30, 30)) 
    
    for i in range(length):
        axs[i][0].imshow(imgs[i]) 
        axs[i][0].axis('off') 
        axs[i][0].set_title('Original ' + labels[i])
        
        axs[i][1].imshow(imgs_gamma[i]) 
        axs[i][1].axis('off') 
        axs[i][1].set_title('Gamma corrected ' + labels[i]) 
        
    plt.show()
```

```python
gamma = 2.0
PATH = "drive/MyDrive/images/"

images =[
    plt.imread(PATH+'low_detalized.jpg'),
    plt.imread(PATH+'high_detalized.jpeg'),
    plt.imread(PATH+'low_contrast.jpg'),
    plt.imread(PATH+'high_contrast.jpg')
]
labels = [
    "low detalized",
    "high detalized",
    "low contrast",
    "high contrast"
    ]

gamma_images=[gamma_increase(img, gamma) for img in images]
plot_images(images, gamma_images, labels)
```
