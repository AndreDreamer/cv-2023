```python
import matplotlib.pyplot as plt
import cv2
```

```python
def left_shift(filename, k=30):
    image = cv2.imread(filename)

    for x in range(0, len(image)):
        for y in range(len(image[1])):
            pixel = image[x][y]
            for color in range(3):
                if ((pixel[color] + k) < 256):
                    pixel[color] = pixel[color] + k
                else:
                    pixel[color] = 255
    return image


def plate_img(img):
    plt.imshow(img)
    plt.title("Shifted")
    plt.axis('off')
    plt.show()


def plate_img_file(filename):
    image = cv2.imread(filename)
    plt.imshow(image)
    plt.title("Original")
    plt.axis('off')
    plt.show()


images = [
    "high-contrast.jpg",
    "high-detailed.jpg",
    "low-contrast.jpg",
    "low-detailed.jpg",
]

for filename in images:
  plate_img_file(filename)
  image = plate_img(left_shift(filename))
```
