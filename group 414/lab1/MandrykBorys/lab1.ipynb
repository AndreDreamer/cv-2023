```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def box_filter(img_data, kernel_size):
    area = kernel_size[0] * kernel_size[1]
    n_rows, n_cols, _ = img_data.shape
    new_img = []

    for r in range(0, n_rows - kernel_size[0] + 1):
        row_data = []
        for c in range(0, n_cols - kernel_size[1] + 1):
            pixel = [0, 0, 0]
            for i in range(r, r + kernel_size[0]):
                for j in range(c, c + kernel_size[1]):
                    pixel[0] += img_data[i, j][0] / area
                    pixel[1] += img_data[i, j][1] / area
                    pixel[2] += img_data[i, j][2] / area
            row_data.append(pixel)
        new_img.append(row_data)
    return np.array(new_img).astype(np.uint8)
```

```python
def plot(input_image, kernel_size):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
    ax[0].imshow(input_image)
    ax[0].set_title('Input Image')
    ax[0].axis('off')

    image = box_filter(input_image, kernel_size)
    ax[1].imshow(image)
    ax[1].set_title(f'Box Filter {kernel_size}')
    ax[1].axis('off')
```

```python
image1 = cv2.cvtColor(cv2.imread('low_detail.jpg'), cv2.COLOR_BGR2RGB)
```

```python
plot(image1, (3, 3))
```

```python
plot(image1, (5, 5))
```
