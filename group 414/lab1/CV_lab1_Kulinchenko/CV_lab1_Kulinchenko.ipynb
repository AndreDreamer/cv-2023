```python
import numpy as np

def convolve(image, kernel):
    image = np.array(image)
    
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape

    padding_height = kernel_height // 2
    padding_width = kernel_width // 2
    padded_image = np.zeros((image_height + 2 * padding_height, image_width + 2 * padding_width))
    padded_image[padding_height:-padding_height, padding_width:-padding_width] = image

    processed_image = np.zeros_like(image)
    for i in range(image_height):
        for j in range(image_width):
            processed_image[i, j] = (kernel * padded_image[i:i+kernel_height, j:j+kernel_width]).sum()

    return processed_image
```

```python
from matplotlib import pyplot as plt

def show(np_image):
    plt.figure()
    plt.imshow(np_image, cmap="gray")
```

```python
def sobel_edge_detection(image):
    sobel_x = 1/8 * np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_y = 1/8 * np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

    show(convolve(image, sobel_x))
    show(convolve(image, sobel_y))
```

```python
def prewitt_edge_detection(image):
    prewitt_x = 1/8 * np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    prewitt_y = 1/8 * np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

    show(convolve(image, prewitt_x))
    show(convolve(image, prewitt_y))
```

```python
sobel_edge_detection(ball) 
prewitt_edge_detection(ball) 

sobel_edge_detection(dog) 
prewitt_edge_detection(dog) 

sobel_edge_detection(fig) 
prewitt_edge_detection(fig) 

sobel_edge_detection(paris) 
prewitt_edge_detection(paris) 

sobel_edge_detection(tower) 
prewitt_edge_detection(tower) 
```
